```python
import matplotlib.pyplot as plt  
import pandas as pd  
import numpy as np  
import os
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as ch

path = os.getcwd() + '/shopping_data.csv'
customer_data = pd.read_csv(path)  
 
data = customer_data.iloc[:, 3:5].values

lamb = lambda x: 1 if x=="Male" else 0
y = [lamb(x) for x in customer_data.Genre]

# print(y)

# plt.figure(1)
# plt.scatter(data[:,0], data[:,1], c=y, cmap='viridis')
# plt.grid()

for linkage in ("ward", "average", "complete", "single"):
    
    clustering = AgglomerativeClustering(linkage=linkage, metric="euclidean")
    clustering.fit(data)
    plt.figure(1)
    plt.scatter(data[:,0],data[:,1], c=clustering.labels_)
    plt.title(linkage)
    plt.show()
    
    plt.figure(2)
    Z = ch.linkage(data, method=linkage, metric='euclidean')
    fig = plt.figure(figsize=(25,10))
    dn = ch.dendrogram(Z)
    plt.show()

    
    
    print("-------------------------------------------------------------------------------------------------------------")
    
```

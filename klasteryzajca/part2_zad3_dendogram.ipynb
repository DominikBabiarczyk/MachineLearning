```python
from sklearn.datasets import make_moons
import seaborn as sns
import pandas as pd
from sklearn.cluster import  KMeans
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
 
x, y = make_moons(1000, noise=.05, random_state=0)
X_moon = pd.DataFrame(x, columns = ['f1','f2'])


#metoda klasteryzacji
# km -model KMeans

for linkage in ("ward", "average", "complete", "single"):
    clustering = AgglomerativeClustering(linkage=linkage, n_clusters=2)
    pred_y = clustering.fit_predict(x)
     
    #wykres
    X_moon['k_means'] = pred_y
    sns.lmplot(data=X_moon, x='f1', y='f2', fit_reg=False, hue = 'k_means', palette = ['#eb6c6a', '#6aeb6c']).set(title='Algorytm k-srednich')
    # plt.scatter(clustering.cluster_centers_[:, 0], clustering.cluster_centers_[:, 1], c='black', s=100, alpha=0.5)
    plt.show()
```
